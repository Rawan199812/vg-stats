```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')
```

## Which company is the most common video game publisher


```python
df ['Publisher'].value_counts().idxmax() # return the row label of the maximum value
```

## What’s the most common platform?


```python
df ['Platform'].value_counts().idxmax()
```

## What about the most common genre?


```python
df ['Genre'].value_counts().idxmax()
```

## What are the top 20 highest grossing games?


```python
df.sort_values('Global_Sales', ascending=False).head(20)
```


## For North American video game sales, what’s the median?
1. Provide a secondary output showing ten games surrounding the median sales output
2. assume that games with same median value are sorted in descending order

```python

mn = df['NA_Sales'].median()

df[(df['NA_Sales'] <= mn+1) & (df['NA_Sales'] >= mn-1)].head(10)
```

## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?


```python
df['NA_Sales'].mean()
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?


```python
w = df[df['Platform'] == 'Wii'].Global_Sales.mean()
p = df[df['Platform'] != 'Wii'].Global_Sales.mean()
print(f'({p}, {w})')
```
